```python
from faster_whisper import WhisperModel

audio_path = "sample-audio_Audio-test.m4a"  # ضع الملف بجانب السكربت
model = WhisperModel("medium", compute_type="int8")  # جرّب "small" لو جهازك ضعيف

segments, info = model.transcribe(audio_path, language="ar")
print("Language:", info.language, "Prob:", info.language_probability)
text = "".join([seg.text for seg in segments])
print("Transcript:", text)
```

```python
# تثبيت الحزمة عند الحاجة
# !pip install edge-tts

import edge_tts

async def tts_ar(text, out="voice.mp3", voice="ar-SA-ZariyahNeural"):
    tts = edge_tts.Communicate(text, voice=voice)
    await tts.save(out)

# لاحظ: في Jupyter/Colab نستخدم await مباشرة
text = ( "انا جميل محجوب وهذا هو أول مشروع لي في تحويل النص إلى كلام باستخدام الذكاء الاصطناعي. "
    "تجربة رائعة وأتطلع لتطوير المزيد من المشاريع المشابهة في المستقبل!")

await tts_ar(text, out="voice.mp3")
print("Saved voice.mp3")
```
